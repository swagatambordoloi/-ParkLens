```python
import cv2
import numpy as np
import json
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ultralytics import YOLO
import paho.mqtt.client as mqtt
from IPython.display import clear_output
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Apply seaborn theme for all upcoming graphs
sns.set_theme(style="whitegrid")

# --- MQTT Configuration ---
BROKER = "broker.emqx.io" # Free public broker for testing
PORT = 1883
TOPIC = "smartparking/virtual_lot_1/status"

client = mqtt.Client()
try:
    client.connect(BROKER, PORT, 60)
    client.loop_start()
    print("Successfully connected to virtual MQTT broker!")
except Exception as e:
    print(f"MQTT Connection failed: {e}")
```

```python
# --- Cell 3: Define Your Virtual Parking Spots (Updated - Larger Boxes) ---
import numpy as np
# Expanded Regions of Interest (ROIs) for your parking spots.
# Format for each point: [x, y]
# Order of points: [Top-Left, Top-Right, Bottom-Right, Bottom-Left]
parking_spots = {#these valueses are taken using parkin_space_segmentation_dynamic.py
  "spot_1": np.array([[854, 449], [1276, 444], [1018, 233], [770, 239]], np.int32),
  "spot_2": np.array([[850, 442], [766, 237], [537, 237], [433, 450]], np.int32),
  "spot_3": np.array([[433, 441], [529, 238], [261, 221], [40, 431]], np.int32)
}

# Dictionary to hold the real-time status (0 = Empty, 1 = Occupied)
spot_status = {spot: 0 for spot in parking_spots.keys()}
```

```python
model = YOLO('yolov8n.pt') 
video_path = "parking_lot_simulation2.mp4" # Replace with your video file
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video file.")

# Extract FPS to calculate real time in seconds
fps = cap.get(cv2.CAP_PROP_FPS)
if fps == 0: 
    fps = 30.0 

frame_count = 0

# --- Data Logging Arrays ---
time_log = []
occupancy_log = []
spot_history = []
confidence_log = []
car_centers = []
background_frame = None

try:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break 
            
        frame_count += 1
        current_time_sec = frame_count / fps
        spot_status = {spot: 0 for spot in parking_spots.keys()}

        # Save background frame for the spatial heatmap later
        if frame_count == 1:
            background_frame = frame.copy()

        results = model(frame, classes=[2], verbose=False) 
        
        frame_confidences = []

        for result in results:
            boxes = result.boxes
            for box in boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                cx, cy = int((x1 + x2) / 2), int((y1 + y2) / 2)
                
                # Log confidence and location
                frame_confidences.append(float(box.conf[0]))
                car_centers.append((cx, cy))
                
                # Check intersections
                for spot_name, polygon in parking_spots.items():
                    if cv2.pointPolygonTest(polygon, (cx, cy), False) >= 0:
                        spot_status[spot_name] = 1 
                        
                cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
                cv2.circle(frame, (cx, cy), 5, (0, 0, 255), -1)

        # Average confidence for the frame
        avg_conf = sum(frame_confidences) / len(frame_confidences) if frame_confidences else 0.0
        
        # --- Update Logs ---
        time_log.append(current_time_sec)
        occupancy_log.append(sum(spot_status.values()))
        spot_history.append(list(spot_status.values()))
        confidence_log.append(avg_conf)

        # Draw Polygons
        for spot_name, polygon in parking_spots.items():
            color = (0, 255, 0) if spot_status[spot_name] == 0 else (0, 0, 255) 
            cv2.polylines(frame, [polygon], isClosed=True, color=color, thickness=2)
            cv2.putText(frame, spot_name, (polygon[0][0], polygon[0][1] - 10), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

        # Publish MQTT Payload every 30 frames
        if frame_count % 30 == 0:
            payload = json.dumps({
                "timestamp": time.strftime("%Y-%m-%d %H:%M:%S"),
                "lot_status": spot_status
            })
            client.publish(TOPIC, payload)
            clear_output(wait=True)
            print(f"Published to {TOPIC}: \n{payload}")

        cv2.imshow("Smart Parking Simulation", frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

except KeyboardInterrupt:
    print("Simulation stopped manually.")

cap.release()
cv2.destroyAllWindows()
print("Simulation complete. Data ready for analytics.")
```

```python
#dynamic 
import time
import numpy as np
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# --- Dynamic Benchmarking Configuration ---
models_config = {
    'YOLOv8-Nano': 'yolov8n.pt',
    'YOLOv8-Small': 'yolov8s.pt',
    'YOLOv8-Medium': 'yolov8m.pt'
}
video_path = "parking_lot_simulation2.mp4"
num_benchmark_frames = 50  # Number of frames to test for speed calculation

models = list(models_config.keys())
accuracy = []
fps = []

print("Starting dynamic model benchmarking on your hardware...")

for model_label, model_file in models_config.items():
    print(f"\nEvaluating {model_label} ({model_file})...")
    
    # Load the model
    model = YOLO(model_file)
    
    # 1. DYNAMIC ACCURACY METRIC (mAP50 on standardized validation set)
    print(f" -> Calculating validation accuracy score...")
    metrics = model.val(data='coco8.yaml', verbose=False)
    accuracy.append(metrics.box.map50)  # Use mAP50 as the standard accuracy score proxy
    
    # 2. DYNAMIC SPEED METRIC (FPS on your video)
    print(f" -> Benchmarking inference speed on your video...")
    cap = cv2.VideoCapture(video_path)
    
    if not cap.isOpened():
        print(f" -> Warning: Could not open {video_path} for speed test. Using synthetic frame.")
        # Fallback to synthetic frame if video isn't reachable
        frame = np.zeros((640, 640, 3), dtype=np.uint8)
        start_time = time.time()
        for _ in range(num_benchmark_frames):
            model(frame, verbose=False)
        end_time = time.time()
    else:
        frames = []
        for _ in range(num_benchmark_frames):
            ret, frame = cap.read()
            if not ret:
                break
            frames.append(frame)
        cap.release()
        
        # Measure elapsed time for inference
        start_time = time.time()
        for frame in frames:
            model(frame, classes=[2], verbose=False)
        end_time = time.time()
        
    elapsed_time = end_time - start_time
    calculated_fps = len(frames) / elapsed_time if elapsed_time > 0 else 0
    fps.append(calculated_fps)

print("\nBenchmarking complete! Generating chart...")

# --- Updated Dynamic Plotting Section ---
fig, ax1 = plt.subplots(figsize=(10, 6))
x = np.arange(len(models))
width = 0.35

# Plot Accuracy Bars (Left Y-Axis)
rects1 = ax1.bar(x - width/2, accuracy, width, label='Accuracy (mAP50)', color='#3475b3')
ax1.set_ylabel('Accuracy Score (mAP50)', fontsize=12, color='#3475b3')
ax1.set_ylim(0, 1.1)

# Plot FPS Bars (Right Y-Axis)
ax2 = ax1.twinx()
rects2 = ax2.bar(x + width/2, fps, width, label='Inference Speed (FPS)', color='#ff7f0e')
ax2.set_ylabel('Frames Per Second (FPS)', fontsize=12, color='#ff7f0e')
# Dynamically adjust the right Y-axis ceiling depending on your hardware's speed
ax2.set_ylim(0, max(fps) * 1.2 if fps else 60)

# Titles, Labels, and Layout Configurations
ax1.set_title('Model Trade-off: Accuracy vs. Edge Inference Speed', fontsize=16, fontweight='bold', pad=15)
ax1.set_xticks(x)
ax1.set_xticklabels(models, fontsize=12)
ax1.bar_label(rects1, fmt='%.2f', padding=3)
ax2.bar_label(rects2, fmt='%.1f', padding=3)

fig.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import json
import time
from ultralytics import YOLO
from IPython.display import clear_output
import sqlite3
# Define the database function right before the loop to guarantee execution
def log_occupancy_to_db(timestamp, occupied_count, raw_statuses):
    """Safely commits real-time tracking metrics to a persistent SQLite database file."""
    conn = sqlite3.connect("parking_metrics.db")
    cursor = conn.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS logs 
        (timestamp TEXT, occupied INTEGER, status_payload TEXT)
    """)
    cursor.execute("INSERT INTO logs VALUES (?, ?, ?)", (timestamp, occupied_count, str(raw_statuses)))
    conn.commit()
    conn.close()

# Initialize Model and Video Stream
model = YOLO('yolov8n.pt') 
video_path = "parking_lot_simulation2.mp4" 
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video file.")

# Extract FPS to calculate real video timeline seconds
fps = cap.get(cv2.CAP_PROP_FPS)
if fps == 0: 
    fps = 30.0 

frame_count = 0

# --- Clear and Reset Data Logging Arrays ---
time_log = []
occupancy_log = []
spot_history = []
confidence_log = []
car_centers = []
background_frame = None

try:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break 
            
        frame_count += 1
        current_time_sec = frame_count / fps
        
        # Reset current frame's status map tracking
        spot_status = {spot: 0 for spot in parking_spots.keys()}

        # Save a background frame for the spatial heatmap analytics later
        if frame_count == 1:
            background_frame = frame.copy()

        # Class 2 corresponds to 'car' in the COCO dataset config
        results = model(frame, classes=[2], verbose=False) 
        frame_confidences = []

        for result in results:
            boxes = result.boxes
            for box in boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                cx, cy = int((x1 + x2) / 2), int((y1 + y2) / 2)
                
                # Log confidence metrics and bounding box centers
                frame_confidences.append(float(box.conf[0]))
                car_centers.append((cx, cy))
                
                # Check intersection using OpenCV's pointPolygonTest geometry math
                for spot_name, polygon in parking_spots.items():
                    if cv2.pointPolygonTest(polygon, (cx, cy), False) >= 0:
                        spot_status[spot_name] = 1 
                        
                cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
                cv2.circle(frame, (cx, cy), 5, (0, 0, 255), -1)

        # Average confidence for the frame calculations
        avg_conf = sum(frame_confidences) / len(frame_confidences) if frame_confidences else 0.0
        current_occupied_count = sum(spot_status.values())
        
        # --- Update Global Real-Time Memory Logs ---
        time_log.append(current_time_sec)
        occupancy_log.append(current_occupied_count)
        spot_history.append(list(spot_status.values()))
        confidence_log.append(avg_conf)

        # 🚀 DATABASE LOGGING STRATEGY: Commit entries every 15 frames (~0.5 seconds)
        if frame_count % 15 == 0:
            current_timestamp = time.strftime("%Y-%m-%d %H:%M:%S")
            log_occupancy_to_db(current_timestamp, current_occupied_count, spot_status)

        # Draw Polygons overlays onto video frame UI
        for spot_name, polygon in parking_spots.items():
            color = (0, 255, 0) if spot_status[spot_name] == 0 else (0, 0, 255) 
            cv2.polylines(frame, [polygon], isClosed=True, color=color, thickness=2)
            cv2.putText(frame, spot_name, (polygon[0][0], polygon[0][1] - 10), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

        # Publish MQTT Payload every 30 frames (~1 second intervals)
        if frame_count % 30 == 0:
            payload = json.dumps({
                "timestamp": time.strftime("%Y-%m-%d %H:%M:%S"),
                "lot_status": spot_status
            })
            client.publish(TOPIC, payload)
            clear_output(wait=True)
            print(f"Published to {TOPIC}: \n{payload}")
            print(f"Database sync successful. Records permanently logged.")

        cv2.imshow("Smart Parking Simulation", frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

except KeyboardInterrupt:
    print("Simulation stopped manually.")

cap.release()
cv2.destroyAllWindows()
print("Simulation complete. Data successfully logged to 'parking_metrics.db'!")
```
